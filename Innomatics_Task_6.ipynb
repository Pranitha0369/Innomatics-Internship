Problem 1 :Smart Parking Lot Management System.


```python
def manage_parking(capacity, logs):
    current = 0
    peak_usage = 0

    for action in logs:
        if action == "IN":
            current += 1
        elif action == "OUT":
            current = max(0, current - 1)


        if current > peak_usage:
            peak_usage = current


    if current >= capacity:
        status = "Parking lot is full!"
    else:
        status = "Available"

    return {
        "Currently Parked": current,
        "Peak Usage": peak_usage,
        "Status": status
    }

# Execution
capacity = 50
vehicle = list(map(str,input().split(",")))
result = manage_parking(capacity, vehicle)

print(f"Currently Parked Vehicles: {result['Currently Parked']}")
print(f"Parking Status: {result['Status']}")
```

Problem 2: Online Food Dekivery Time Estimator.


```python
def delivery_time(distance,weather,traffic_level,prepare=5):
  base=distance*2
  traffic_delays = {
        "Low": 5,
        "Medium": 15,
        "High": 25
    }

    # Weather Adjustments
  weather_delays = {
        "Clear": 0,
        "Rainy": 10,
        "Snowy": 20
    }
  if traffic_level in traffic_delays:
    traffic=traffic_delays[traffic_level]
  if weather in weather_delays:
    weatherr=weather_delays[weather]
  delivery=base+traffic+weatherr +prepare
  return delivery
distance=int(input("Distance(km):"))
traffic_level=input("Traffic level(Low,High,Medium):")
weather=input("Weather(Clear,Rainy,Snowy):")
print(f"Estimated Delivery Time :{delivery_time(distance,weather,traffic_level)}")

```

Problem 3:Movie Theatre Seat occupancy Analyzer.

```python
def seat_booking(total,booked_seats):
  occupancy = len(booked_seats)
  print(occupancy)
  occupancy_p= (occupancy/total)*100
  print(f"Occupancy : {int(occupancy_p)}%")
  if occupancy_p >=75 and occupancy_p < 98:
    print("Almost full")
  elif occupancy_p <75:
    print("Seats available")
  else:
    print("Housefull!")
total=int(input())
booked_seats = list(map(int,input().split(",")))
seat_booking(total,booked_seats)
```

Problem 4:Cloud Server Load Classification System

```python
def server_load(avg):
  if avg <50:
    return "Normal"
  elif avg >=50 and avg <= 80:
    return "Warning"
  else:
    return "Critical"
cpu=list(map(int,input().split(",")))
avg=sum(cpu)/len(cpu)
print(f"Average CPU LOad :{int(avg)}%")
print(f"Server Status :{server_load(avg)}")
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_resources(usage_data):
  thresholds = {
        "Projector": 5,
        "AC": 8,
        "Lights": 10
    }
  overused_list = []
  energy_alert = "No"
  for resource, hours in usage_data.items():
    limit = thresholds.get(resource, 8)
    if hours > limit:
        overused_list.append(resource)
        energy_alert = "Yes"
  overused_str = ", ".join(overused_list) if overused_list else "None"
  return {
        "Overused Resources": overused_str,
        "Energy Alert": energy_alert
    }
current_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

result = classroom_resources(current_usage)

print(f"Overused Resources: {result['Overused Resources']}")
print(f"Energy Alert: {result['Energy Alert']}")
```

Problem 6:
Online Event Registration Capacity Controller


```python
def manage_event(capacity,registration):
  if registration < capacity :
    print(f"Confirmed Registration :{registration}")
    print(f"Waitlisted Users : 0 ")
    print(f"Registration Status :Open")
  elif registration > capacity :
    print(f"Confirmed Registration :{capacity}")
    print(f"Waitlisted Users : {registration - capacity} ")
    print(f"Registration Status :Closed")
capacity = int(input())
registration = int(input())
manage_event(capacity,registration)
```

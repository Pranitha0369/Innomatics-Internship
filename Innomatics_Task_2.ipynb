Problem 1 : Unique words in a sentence.

```python
ln=input("Enter Sentence : ")
l=ln.split()
unique = set(l)
print(f"Unique words count :{len(unique)}")
print(f"Unique words :{unique}")
```

Problem 2:Highest salary from Employee Dataset


```python
employees = {
    "Ravi" : 75000,
    "Ankita":68000,
    "Kiran":72000
}

highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

print(f"All employee names: {list(employees.keys())}")
print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

Problem 3: Find MAximum and Minimum Values


```python
numbers=list(map(int,input().split(",")))

if numbers:
  max_val = numbers[0]
  min_val = numbers[0]

  for num in numbers:
    if min_val > num:
      min_val=num
    elif max_val < num:
      max_val=num
  print(f"Max :{max_val}")
  print(f"Min :{min_val}")
else:
  print("The list of numbers is empty.")
```

Problem 4: Count Products above a Price threshold


```python
prices = list(map(int,input().split(",")))
threshold=1000
l=[]
for price in prices:
  if price > threshold:
    l.append(price)
print(f"Products above {threshold} :{len(l)}")
```

Problem 5: Calculate Attendance Percentage

```python
attendance = list(map(str,input("Only P and A:").split(",")))
prsent=attendance.count("P")
percentage=(prsent/len(attendance))*100
print(f"Attendance percentage :{percentage}")
```

Problem 6 : Remove Duplicates Phone Number


```python
numbers=list(map(int,input().split(",")))
unique_numbers =set(numbers)
print(f"Unique phone numbers :{unique_numbers}")
```

Problem 7 :Count Character Frequency


```python
s=input()
d={}
for char in s:
  if char in d:
    d[char]+=1
  else:
    d[char]=1
print(d)
```

Problem 1 :Mobile Recharge VAlidation System


```python
def validate():
  valid=[199,299,399,599]
  while True:
    amount=int(input("Enter the amount :"))
    if amount >= 50 and amount in valid :
      print(f"Recharge successfull")
      break
    elif amount <50 :
      print(f"The valid recharge plans are :{valid}")
    else:
      print(f"Recharge failed")
validate()
```

Problem 2:Inventory Reorder Alert system


```python
def reorder(k,v):
  if v <15 :
    print(f" Please,{k} reorder the stock ")
  else :
    print(f" Inventory {k} Stock Ok")
inventory ={}
n=int(input("Enter total number of inventories :"))
for k in range(n):
  k=input("Enter inventory name :")
  val=int(input("enter stock quantities :"))
  inventory[k]=val
for k,v in inventory.items():
  reorder(k,v)
```

Problem 3: Student Result Processing System


```python
def evaluation():
  total=int(input("Enter total number of students marks :"))
  l=[]
  for i in range(total):
    mark=int(input("Enter each subject marks :"))
    l.append(mark)
  avg=sum(l)//len(l)
  print(f"Average :{avg}")
  if avg >=50 :
    print("Pass")
  else:
    print("Fail")
evaluation()

```

Problem 4: CAb Fare estimator with retry option


```python
def fare_calculation():
  base_fare = 50
  rate = 12
  peak_hour = 1.25
  distance= int(input("Enter distance :"))
  ph=input("Enter is it peak hour or not(yes or no):")
  if ph == "no":
    total_fare = base_fare+(rate*distance)
    print(f"The total cab fare is :{total_fare}")
  else :
    total_fare = base_fare+(rate*distance)*peak_hour
    print(f"The total cab fare is :{total_fare} ")
fare_calculation()
```

Problem 5 : Employee Attendance Eligibility Checker


```python
def attendance():
  attendee = list(map(str,input("Enter Attendance :").split(",")))
  presentee=attendee.count("P")
  total = len(attendee)
  attend = (presentee/total)*100
  return attend
result = attendance()
print(f"Total attendance Percentage : {result}%")
if result >=75 :
  print("Eligible")
else:
  print("Not Eligibile")
```

Problem 6:Password Strength Checker


```python
def password(passwd):
  if len(passwd) >=8:
    if passwd.isdigit() == False :
      print("Please enter atleast one digit")
    elif passwd.isspelled() == False :
      print("Must involve atleast one special character")
  else:
    print("Not a strong password(length should be minimum 8)")
passwd = input("Enter password :")
password(passwd)
```
